```python
import numpy as np 
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#.1    )plot scatter )  horsepower  mpg
#     origin(  (
df = pd.read_csv("mpg.csv")
sns.scatterplot(data=df, x='mpg', y='horsepower',hue ='mpg')
plt.show()
```

```python
#.2     4       8 
#)  mpg),      )chart bar )

filtered_df = df[df['cylinders'].isin([4, 8])]
avg_mpg = filtered_df.groupby('cylinders', as_index=False)['mpg'].mean()

sns.barplot(data=avg_mpg, x='cylinders', y='mpg',hue='mpg')

plt.show()
```

```python
#.3    )plot count ) origin        
#

sns.countplot(data=df,x='origin')
```

```python
#.4   )histogram ) weight       
sns.displot(data = df , x='weight' , bins = 20)
```

```python
#.5  plot box   mpg    ) cylinders )

sns.boxplot(data=df, x='cylinders', y='mpg')

plt.show()
```
