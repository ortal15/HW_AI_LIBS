```python
import numpy as np 
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#.1 צור תרשים פיזור )plot scatter )בין עמודת horsepower לעמודת mpg
#הוסף צבעים שונים לפי עמודת origin( מקור הרכב(
df = pd.read_csv("mpg.csv")
sns.scatterplot(data=df, x='mpg', y='horsepower',hue ='mpg')
plt.show()
```

```python
#.2 בדוק אם רכבים עם 4 צילינדרים צורכים פחות דלק מרכבים עם 8 צילינדרים
#)באמצעות עמודת mpg), הצג זאת באמצעות תרשים עמודות )chart bar )

filtered_df = df[df['cylinders'].isin([4, 8])]
avg_mpg = filtered_df.groupby('cylinders', as_index=False)['mpg'].mean()

sns.barplot(data=avg_mpg, x='cylinders', y='mpg',hue='mpg')

plt.show()
```

```python
#.3 צור תרשים ספירה )plot count )לעמודת origin כדי לבדוק מאיזו מדינה יש הכי הרבה רכבים
#בקובץ

sns.countplot(data=df,x='origin')
```

```python
#.4 צור היסטוגרמה )histogram )לעמודת weight כדי להציג את התפלגות המשקל של הרכבים
sns.displot(data = df , x='weight' , bins = 20)
```

```python
#.5 צור plot box להשוואת ערכי mpg לפי מספר הצילינדרים ) cylinders )

sns.boxplot(data=df, x='cylinders', y='mpg')

plt.show()
```
